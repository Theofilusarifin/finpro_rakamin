### Numerical Data Preprocessing

#### Removing Outliers

```python
def remove_outliers_iqr(dataframe, column_name):
    # Calculate the IQR
    Q1 = dataframe[column_name].quantile(0.25)
    Q3 = dataframe[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter the DataFrame to exclude outliers
    filtered_dataframe = dataframe[(dataframe[column_name] >= lower_bound) & (
        dataframe[column_name] <= upper_bound)]

    # Display information about the removed outliers
    outliers_removed = dataframe.shape[0] - filtered_dataframe.shape[0]
    print(f"Number of outliers removed in '{column_name}': {outliers_removed}")

    return filtered_dataframe
```

```python
for i in range(0, len(nums)):
    df = remove_outliers_iqr(df, nums[i])
```

## Clustering

### Elbow Method

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import metrics
```

```python
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)
```

```python
# Plot elbow method
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal Cluster Number')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

### Clustering

```python
optimal_cluster_number = 3
kmeans = KMeans(n_clusters=optimal_cluster_number, random_state=42, n_init=10)
labels = kmeans.fit_predict(df_scaled)
```

```python
df_scaled['Cluster_Labels'] = labels
df_scaled
```

```python
df_scaled.to_csv('../dataset/result.csv')
```

### PCA

```python
pca = PCA(n_components=2)
data_pca = pca.fit_transform(df_scaled)
```

```python
df_pca = pd.DataFrame(data_pca, columns=['PC1', 'PC2'])
df_pca['Cluster'] = labels
```

```python
plt.figure(figsize=(10, 7))
for cluster in range(optimal_cluster_number):
    cluster_data = df_pca[df_pca['Cluster'] == cluster]
    plt.scatter(cluster_data['PC1'], cluster_data['PC2'],
                label=f'Cluster {cluster}')

plt.title('Clustering Result Visualization with PCA')
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.legend()
plt.show()
```
